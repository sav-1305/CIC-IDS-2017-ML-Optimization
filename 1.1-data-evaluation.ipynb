```python
import pandas as pd
```

```python
df = pd.read_csv(r"1.0-Output/AllDaysCombined.csv")
```

```python
non_feature_columns = ['id', 'Flow ID','Src IP','Dst IP','Timestamp', 'Label', 'Attempted Category']
df.drop(columns=non_feature_columns, inplace=True)
del non_feature_columns
```

```python
import numpy as np

inf_counts = df.replace([np.inf, -np.inf], np.nan).isnull().sum() - df.isnull().sum()

print(f"Infinity Counts:\n{inf_counts[inf_counts > 0].sort_values(ascending=False)}\n\n")

large_vals = df.apply(lambda col: (np.abs(col) > 1e10).sum())
print(f"Large Value Counts:\n{large_vals[large_vals > 0].sort_values(ascending=False)}\n\n")

df.replace([np.inf, -np.inf], np.nan, inplace=True)

df.dropna(inplace=True)

print("Remaining infinities:", np.isinf(df.to_numpy()).sum())
print("Remaining NaNs:", df.isnull().sum().sum())

del inf_counts
del large_vals
```

```python
from pathlib import Path

# Define the folder and ensure it exists
output_dir = Path("1.1-Output")
output_dir.mkdir(parents=True, exist_ok=True)

df.to_csv(r"1.1-Output/ProcessedFeatures.csv", index=False)
```
